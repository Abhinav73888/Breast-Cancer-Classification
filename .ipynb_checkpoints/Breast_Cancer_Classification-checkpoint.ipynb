## Importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 

warnings.filterwarnings('ignore')
```

## Reading dataset

```python
data = pd.read_csv('./data/data.csv')
data.head()
```

## Checking whether data is balanced or not

```python
sns.countplot(data['diagnosis'])
plt.show()
```

## Data cleaning

```python
# dropping some unnecessary columns
data.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
# checking for null values
data.isna().sum()
```

**Observing the above distribution plots, some of the important features can be selected.**

```python
reduced_data = data[['diagnosis', 'radius_mean', 'area_mean', 'concave points_mean', 'perimeter_mean', 'concavity_mean', 'area_se', 'perimeter_worst', 'radius_worst', 'concavity_worst', 'concave points_worst']]
```

## Plotting the correlation map

```python
plt.figure(figsize=(12,10))
sns.heatmap(reduced_data.corr(), annot=True, cmap='Blues')
```

## Pair plot among highly correlated features

```python
sns.pairplot(reduced_data[['radius_mean', 'area_mean', 'perimeter_mean']])
```

**Radius mean, area mean and perimeter mean are highly correlated so we can select just one among them.**

```python
reduced_data.drop(columns=['area_mean', 'perimeter_mean'], inplace=True)
```

## Data preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

# encoding categorical to numeric
enc = LabelEncoder()
reduced_data['diagnosis'] = enc.fit_transform(reduced_data['diagnosis'])

# splitting data into features and labels
features = reduced_data.drop(columns=['diagnosis'])
labels = reduced_data['diagnosis']

# scaling features as SVM is range sensitive 
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# splitting data into train, test with ratio of 80-20
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42) 
```

## Applying SVM model for classification

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(x_train, y_train)
```

## Model evaluation

```python
y_train_pred = clf.predict(x_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, plot_confusion_matrix, plot_roc_curve
```

```python
print('Train_accuracy:', accuracy_score(y_train,y_train_pred))
print('Test_accuracy:', accuracy_score(y_test,y_pred))
```

```python
plot_confusion_matrix(clf, x_train, y_train, cmap='Blues')
```

```python
plot_confusion_matrix(clf, x_test, y_test, cmap='Blues')
```

```python
plot_roc_curve(clf, x_test, y_test)
```
